````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
N = 100
x_arr = []
y_arr = []
dx = 2*np.pi/N

for i in range(N):
	x = i*dx
	y = np.sin(x)
	x_arr.append(x)
	y_arr.append(y)

plt.plot(x_arr, y_arr)
plt.show()
````

````python
x = np.random.normal(0, 1, 10)  #  0,  1  10  
y = np.random.normal(0, 1, 10)

size = np.random.uniform(10, 100, 10)  # 10 100  10  .
colors = np.random.uniform(0, 1, 10)  # 0 1    

plt.scatter(x, y, s=size, c=colors)
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar()
plt.show()
````

````python
x = np.random.normal(0, 1, 10000)

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.hist(x, bins=50, edgecolor="white")
plt.title("frequency distribution")

plt.subplot(1,3,2)
plt.hist(x, bins=50, edgecolor="white", density=True)
plt.title("frequency distribution ratio")

plt.subplot(1,3,3)
plt.hist(x, bins=50, edgecolor="white", cumulative=True)
plt.title("cumulative distribution")
plt.show()
````

````python
wave_length = 15.0  #  15
k = np.pi*2/wave_length  # 
side = 100  #    100
points = 500  #   500 
spacing = side/points  #   500   
A = np.empty([points, points], float)  #    
````

````python
for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*j
        r = np.sqrt(x**2 + y**2)  #  
        A[i, j] = np.sin(k*r)  #  
````

````python
plt.imshow(A, origin="lower", extent=[0, side, 0, side])  #   
plt.xlabel("x (cm)")
plt.ylabel("y (cm)")
plt.colorbar()
plt.viridis()
plt.show()
````

````python
wave_length = 5.0  #  15
k = np.pi*2/wave_length  # 
side = 100  #    100
points = 500  #   500 
spacing = side/points  #   500   
A = np.empty([points, points], float)  #    
B = np.empty([points, points], float)  #    
C = np.empty([points, points], float)  # A+B   

for i in range(points):
    y = spacing * (i - points / 2)  #   y 
    for j in range(points):
        x = spacing * (j - points / 2)-15  #   x 
        r = np.sqrt(x**2 + y**2)  #  
        A[i, j] = np.sin(k * r)  #  

for i in range(points):
    y = spacing * (i - points / 2)  #   y 
    for j in range(points):
        x = spacing * (j - points / 2)+15  #   x 
        r = np.sqrt(x**2 + y**2)  #  
        B[i, j] = np.sin(k * r)  #  

C = A+B

plt.figure(figsize=(15, 4))

plt.subplot(1,3,1)
plt.imshow(A, origin="lower", extent=[-side / 2, side / 2, -side / 2, side / 2])  #   
plt.xlabel("x (cm)")
plt.ylabel("y (cm)")
plt.colorbar()
plt.viridis()

plt.subplot(1,3,2)
plt.imshow(B, origin="lower", extent=[-side / 2, side / 2, -side / 2, side / 2])  #   
plt.xlabel("x (cm)")
plt.ylabel("y (cm)")
plt.colorbar()
plt.viridis()

plt.subplot(1,3,3)
plt.imshow(C, origin="lower", extent=[-side / 2, side / 2, -side / 2, side / 2])  #   
plt.xlabel("x (cm)")
plt.ylabel("y (cm)")
plt.colorbar()
plt.viridis()
plt.show()
````
