````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
N = 100
x_arr = []
y_arr = []
dx = 2*np.pi/N

for i in range(N):
	x = i*dx
	y = np.sin(x)
	x_arr.append(x)
	y_arr.append(y)

plt.plot(x_arr, y_arr)
plt.show()
````

````python
wave_length = 15.0  # 파장 15
k = np.pi*2/wave_length  # 파수
side = 100  # 한 변의 길이 100
points = 500  # 한 변을 500개로 나눌것임
spacing = side/points  # 한 변을 500으로 나눠 간격을 계산함
A = np.empty([points, points], float)  # 데이터를 기록할 빈 행렬
````

````python
for i in range(points):
    y = spacing*i
    for j in range(points):
        x = spacing*j
        r = np.sqrt(x**2 + y**2)  # 원점으로부터의 거리
        A[i, j] = np.sin(k*r)  # 원형 파형
````

````python
plt.imshow(A, origin="lower", extent=[0, side, 0, side])  # 원점을 왼쪽 아래에
plt.xlabel("x (cm)")
plt.ylabel("y (cm)")
plt.colorbar()
plt.viridis()
plt.show()
````

```{python}
plt.imshow(A, origin="lower", extent=[xmin, xmax, ymin, ymax])
```
+ origin="lower": 이미지의 원점을 왼쪽 아래로 지정합니다. 이는 행렬의 표현 방식과 관련이 있습니다. 행렬에서 첫 번째 행이 아래쪽에 있기 때문에 일반적으로 이미지 데이터를 처리할 때 사용합니다.

+ origin="upper": 이미지의 원점을 왼쪽 위로 지정합니다. 이는 일반적인 좌표 체계에서 사용되는 방식입니다.

+ 첫 번째 매개변수는 표시하려는 이미지 데이터입니다. 이 데이터는 2D 배열 또는 행렬이어야 합니다. imshow는 이러한 데이터를 색상으로 매핑하여 이미지로 표시합니다.

````python
for i in range(points):
    y = spacing * (i - points / 2)  # 중심을 기준으로 좌표 이동
    for j in range(points):
        x = spacing * (j - points / 2)  # 중심을 기준으로 좌표 이동
        r = np.sqrt(x**2 + y**2)  # 원점으로부터의 거리
        A[i, j] = np.sin(k * r)  # 원형 파형

plt.imshow(A, origin="lower", extent=[-side / 2, side / 2, -side / 2, side / 2])  # 중심을 왼쪽 아래에
plt.xlabel("x (cm)")
plt.ylabel("y (cm)")
plt.colorbar()
plt.viridis()
plt.show()
````

````python
wave_length = 5.0  # 파장 15
k = np.pi*2/wave_length  # 파수
side = 100  # 한 변의 길이 100
points = 500  # 한 변을 500개로 나눌것임
spacing = side/points  # 한 변을 500으로 나눠 간격을 계산함
A = np.empty([points, points], float)  # 데이터를 기록할 빈 행렬
B = np.empty([points, points], float)  # 데이터를 기록할 빈 행렬
C = np.empty([points, points], float)  # A+B를 기록할 빈 행렬

for i in range(points):
    y = spacing * (i - points / 2)  # 중심을 기준으로 y좌표 이동
    for j in range(points):
        x = spacing * (j - points / 2)-15  # 중심을 기준으로 x좌표 이동
        r = np.sqrt(x**2 + y**2)  # 원점으로부터의 거리
        A[i, j] = np.sin(k * r)  # 원형 파형

for i in range(points):
    y = spacing * (i - points / 2)  # 중심을 기준으로 y좌표 이동
    for j in range(points):
        x = spacing * (j - points / 2)+15  # 중심을 기준으로 x좌표 이동
        r = np.sqrt(x**2 + y**2)  # 원점으로부터의 거리
        B[i, j] = np.sin(k * r)  # 원형 파형

C = A+B

plt.figure(figsize=(15, 4))

plt.subplot(1,3,1)
plt.imshow(A, origin="lower", extent=[-side / 2, side / 2, -side / 2, side / 2])  # 중심을 왼쪽 아래에
plt.xlabel("x (cm)")
plt.ylabel("y (cm)")
plt.colorbar()
plt.viridis()

plt.subplot(1,3,2)
plt.imshow(B, origin="lower", extent=[-side / 2, side / 2, -side / 2, side / 2])  # 중심을 왼쪽 아래에
plt.xlabel("x (cm)")
plt.ylabel("y (cm)")
plt.colorbar()
plt.viridis()

plt.subplot(1,3,3)
plt.imshow(C, origin="lower", extent=[-side / 2, side / 2, -side / 2, side / 2])  # 중심을 왼쪽 아래에
plt.xlabel("x (cm)")
plt.ylabel("y (cm)")
plt.colorbar()
plt.viridis()
plt.show()
````
