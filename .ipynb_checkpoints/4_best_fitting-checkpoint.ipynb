```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hubble_data = np.loadtxt("./data/hubble.txt")
x = hubble_data[:, 0]
y = hubble_data[:, 1]
xx = np.sum(x*x)
xy = np.sum(x*y)
H = xy/xx
print(H)
```

```python
plt.plot(x, y, "x")
plt.plot(x, x*H)
plt.xlabel("d")
plt.ylabel("V")
plt.show()
```

최소제곱법 풀이

```python
# 파일에서 데이터 읽어오기
data = np.loadtxt("./data/fish.txt")  # 파일 형식에 따라 delimiter를 조절

# 데이터 처리
L = data[:, 0]
M = data[:, 1]
x = np.log(L)
y = np.log(M)

# Least-Squares Method를 사용한 최적화 함수 정의
def least_squares_method(x, y):
    n = len(x)
    X = np.column_stack((np.ones(n), x))  # Design matrix
    Y = y.reshape(-1, 1)  # Response variable

    # 최적의 매개변수 계산
    params = np.linalg.inv(X.T @ X) @ X.T @ Y

    a_optimal, b_optimal = params.flatten()
    return a_optimal, b_optimal

# 최적의 매개변수 찾기
a_optimal, b_optimal = least_squares_method(x, y)

# 결과 출력
A_optimal = np.exp(a_optimal)
print(f"Optimal values: a = {a_optimal}, b = {b_optimal}, A = {A_optimal}")

# 예측값 계산
y_pred = a_optimal + b_optimal * x

# 데이터와 예측값 비교 그래프
plt.scatter(x, y, label="Original Data")
plt.plot(x, y_pred, color='red', label="Predicted Data")
plt.xlabel("ln(L)")
plt.ylabel("ln(M)")
plt.legend()
plt.show()
```

경사하강법 풀이

```python
# 파일에서 데이터 읽어오기
data = np.loadtxt("./data/fish.txt")  # 파일 형식에 따라 delimiter를 조절

# 데이터 처리
L = data[:, 0]
M = data[:, 1]
x = np.log(L)
y = np.log(M)

# 경사 하강법을 사용한 최적화 함수 정의
def gradient_descent(x, y, learning_rate=0.01, epochs=50000):
    a = 0.0
    b = 0.0
    n = len(x)

    for epoch in range(epochs):
        y_pred = a + b * x
        error = y_pred - y

        # 편미분 계산
        gradient_a = 2 * np.sum(error)
        gradient_b = 2 * np.sum(x * error)

        # 매개변수 업데이트
        a = a - learning_rate * gradient_a / n
        b = b - learning_rate * gradient_b / n

    return a, b

# 최적의 매개변수 찾기
a_optimal, b_optimal = gradient_descent(x, y)

# 결과 출력
A_optimal = np.exp(a_optimal)
print(f"Optimal values: a = {a_optimal}, b = {b_optimal}, A = {A_optimal}")

# 예측값 계산
y_pred_gd = a_optimal + b_optimal * x

# 데이터와 예측값 비교 그래프
plt.scatter(x, y, label="Original Data")
plt.plot(x, y_pred_gd, color='green', label="Predicted Data (Gradient Descent)")
plt.xlabel("ln(L)")
plt.ylabel("ln(M)")
plt.legend()
plt.show()
```
